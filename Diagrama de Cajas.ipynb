```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Función para calcular los cuartiles e IQR
def calcular_cuartiles_y_iqr(datos):
    # Ordenar los datos
    datos_ordenados = np.sort(datos)

    # Calcular cuartiles
    q1 = np.percentile(datos_ordenados, 25)
    q2 = np.percentile(datos_ordenados, 50) # Mediana
    q3 = np.percentile(datos_ordenados, 75)

    # Calcular el rango intercuartílico (IQR)
    iqr = q3 - q1

    return q1, q2, q3, iqr

# Función que imprime resultados
def imprimir_resultados(datos, q1, q2, q3):
  print("Datos: " + str(np.sort(datos)))
  print(f"Q1: {q1}")
  print(f"Q2 (Mediana): {q2}")
  print(f"Q3: {q3}")
  print(f"IQR: {iqr}")
  print(f"Valores atípicos por debajo de: {q1-1.5*iqr}")
  print(f"Valores atípicos por encima de: {q3+1.5*iqr}")
  print("")

data1 = np.array([1, 2, 5, 9, 10, 10, 11, 12, 13, 13, 14, 19, 25, 30])
data2 = np.append(data1, [3, 4])
data3 = np.append(data2, [26, 27])

q1, q2, q3, iqr = calcular_cuartiles_y_iqr(data1)
imprimir_resultados(data1, q1,q2, q3)

q1, q2, q3, iqr = calcular_cuartiles_y_iqr(data2)
imprimir_resultados(data2, q1,q2, q3)

q1, q2, q3, iqr = calcular_cuartiles_y_iqr(data3)
imprimir_resultados(data3, q1,q2, q3)

data = [data1, data2, data3]

plt.boxplot(data)
plt.title("Diagrama de Caja y Bigotes")
plt.show()

```

```python
import matplotlib.pyplot as plt

def calculate_boxplot(numbers):
    # Calculate Q1, Q2 and Q3
    numbers_sorted = sorted(numbers)
    n = len(numbers)
    q1 = numbers[0] if n < 4 else numbers[(n-1)//4]
    q3 = numbers[n-1] if n % 4 != 0 else numbers[(n//4)-1]
    iqr = q3 - q1

    # Create the boxplot
    plt.boxplot(numbers)

    # Add labels and title
    plt.xticks([i+1 for i in range(len(numbers))], numbers)
    plt.ylabel('Valores')
    plt.title('Diagrama de Cajas')

    # Show the plot
    plt.show()

# Example usage:
numbers = [3, 14, 15, 16, 7, 2, 17, 9, 8, 6]
calculate_boxplot(numbers)
```
